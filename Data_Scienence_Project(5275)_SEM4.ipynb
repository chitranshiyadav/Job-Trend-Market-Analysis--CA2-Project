**3. Data Cleaning**

```
# 21. Convert Joining_Date into datetime format.
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], errors='coerce')
df["Joining_Date"]
```

```
# 22.Standardize Employment Status values (Full-time / Full-Time).
df["Employment_Status"] = df["Employment_Status"].replace({
    "Full-Time":"Full-time",
    "full time":"Full-time"
})
df["Employment_Status"]
```

```
# 23.Rename salary columns for clarity.
df = df.rename(columns={
    "Min_Salary":"Minimum_Salary",
    "Max_Salary":"Maximum_Salary"
})
df
```

```
# 24. Detect and handle unrealistic salary values.
df = df[df["Maximum_Salary"] < 500000]
df
```

```
# 25. Remove duplicate job records if any.
df.drop_duplicates(inplace=True)
df
```

**SECTION C – FEATURE ENGINEERING**

**4. Derived Columns**

```
# 26. Create Average_Salary using Min and Max salary.
df["Average_Salary"] = (df["Minimum_Salary"] + df["Maximum_Salary"]) / 2
print(df["Average_Salary"])
```

```
# 27. Create Experience_Level: (Junior (≤5 years), Mid-Level (6–8 years), Senior (>8 years))
def exp_level(x):
    if x <= 5:
        return "Junior"
    elif x <= 8:
        return "Mid-Level"
    else:
        return "Senior"

df["Experience_Level"] = df["Experience(in Year)"].apply(exp_level)
df["Experience_Level"]
```

```
# 28. Create Salary_Category( Low (<₹50,000), Medium (₹50,000–₹1,00,000), High (>₹1,00,000))
def sal_cat(x):
    if x < 50000:
        return "Low"
    elif x <= 100000:
        return "Medium"
    else:
        return "High"

df["Salary_Category"] = df["Average_Salary"].apply(sal_cat)
print(df["Salary_Category"])
```

```
#29. Create Demand_Score combining Industry Demand and Growth Rate.
df["Demand_Score"] = df["Growth_Rate(1-10)"] * 0.6 + \
                     df["Skill_Rate(1-5)"] * 0.4
df["Demand_Score"]
```

```
# 30. Create Job_Type_Category based on Employment Status.
df["Job_Type_Category"] =df["Employment_Status"]
df["Job_Type_Category"]
```
